```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dados = pd.read_csv('/Cópia de Dados_Imersão [faça uma cópia para editar] - Dados.csv')
avaliacoes = pd.read_csv('/Dados_Imersão [faça uma cópia para editar] - dados (1).csv')

```

```python
# Remover o símbolo da moeda da coluna "valor_rupias"
dados['actual_price'] = dados['actual_price'].str.replace('₹', '', regex=True)


```

```python
dados['actual_price'] = dados['actual_price'].str.replace('.', '', regex=True)
```

```python
dados['actual_price'] = dados['actual_price'].str.replace(',', '.', regex=True)
```

```python
# Converter a coluna para o tipo numérico (opcional)
dados['actual_price'] = pd.to_numeric(dados['actual_price'])
```

```python
dados.head()

```

```python
# Remover o símbolo da moeda da coluna "product_price"
dados['product_price'] = dados['product_price'].str.replace('₹', '', regex=True).str.replace('.', '', regex=True).str.replace(',', '.', regex=True)

# Converter a coluna para o tipo numérico (float)
dados['product_price'] = pd.to_numeric(dados['product_price'])
# Exibir as primeiras linhas do DataFrame para verificar as alterações
dados.head()
```

```python
dados['rating'] = dados['rating'].str.replace(',','.')
```

```python
dados ['rating'] = pd.to_numeric(dados['rating'])
```

```python
dados.info()
```

```python
dados.describe()
```

###Graficos


```python
 ratings_above_4_5 = dados[dados['rating']>4.5]
 rating_counts= ratings_above_4_5['rating'].value_counts().sort_index()

 plt.figure(figsize=(10,6))
 sns.barplot(x=rating_counts.index, y=rating_counts.values)
 plt.xlabel('Contagem')
 plt.ylabel('Avaliações')
 plt.title('Contagem de Avaliações com Rating > 4.5')
 plt.show()
```

```python
# Fazer os filtros
faixa1 = dados[dados['actual_price'] <= 50]
faixa2 = dados[(dados['actual_price'] > 50) & (dados['actual_price'] <= 500)]
faixa3 = dados[dados['actual_price'] > 500]

# Contar os produtos em cada faixa
contagem_faixa1 = len(faixa1)
contagem_faixa2 = len(faixa2)
contagem_faixa3 = len(faixa3)

# Criar um gráfico de pizza
labels = ['Até 50 reais', 'Entre 50 e 500 reais', 'Mais de 500 reais']
sizes = [contagem_faixa1, contagem_faixa2, contagem_faixa3]
explode = (0.1, 0, 0)  # Explodir a fatia maior (destacar)

plt.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%', shadow=True, startangle=140)
plt.axis('equal')  # Garantir que o gráfico seja um círculo

plt.title('Contagem de Produtos por Faixa de Valor em Reais')
plt.show()
```
